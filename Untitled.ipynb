```python
from konlpy.tag import Okt  
okt=Okt()
f=open(r'C:\Users\IBK\Desktop\fluid_engineering.txt',encoding="utf8")

text=[]
while True:
    line=f.readline()
    if okt.nouns(line)!=[]:
        text.append(okt.nouns(line))
    if not line:
        break
print(text)
f.close()
```

```python
print(len(text))
```

```python
from gensim.models import Word2Vec
model = Word2Vec(text,size=300,window=5,min_count=5,workers=4,sg=0)
# CBOW 방식으로 학습완료
```

```python
from gensim.models import Word2Vec
model = Word2Vec(text,size=200,window=5,min_count=5,workers=4,sg=1)
# Skip-gram 방식으로 학습완료 (전반적으로 CBOW보다 좋은 성능)
```

```python
model_result1=model.wv.most_similar("가속도")
print(model_result1)
```

모든 단어를 형태소 분석한 후, 명사 부분에서 틀린 부분만 고친 후, str 타입으로 변환하는 방식.

```python
from konlpy.tag import Okt  
okt=Okt()
f=open(r'C:\Users\IBK\Desktop\fluid_engineering.txt',encoding="utf8")

text_pos=[]
while True:
    line=f.readline()
    if okt.pos(line)!=[]:
        text_pos.append(okt.pos(line))
    if not line:
        break
print(text_pos)
f.close()
```

```python
text_pos_list=sum(text_pos,[])
#print(text_pos_list) 메모리 부족으로 오류생김.
```
